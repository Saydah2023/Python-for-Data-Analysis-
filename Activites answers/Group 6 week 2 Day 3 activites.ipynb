```python
#1.Write a Python program to test if a given page is found or not on the server.

from bs4 import BeautifulSoup
import requests

html = urlopen('https://www.google.ca/')

url_obj = BeautifulSoup(html,"html5lib")
if url_obj.find("title"):
   print("Page found sucessfully on server")
else:
   print("Page not found")


```

```python
#2.Write a Python program to extract and display all the header tags from en.wikipedia.org/wiki/Main_Page
from urllib.request import urlopen
html = urlopen('https://en.wikipedia.org/wiki/Main_Page')
header_obj = BeautifulSoup(html,"html5lib")
title = header_obj.find_all(['h1','h2','h3','h4','h5','h6'])
print("list all the header tags:",*title,sep='\n\n')
```

```python
#3.	Write a Python program to get 90 days of visits broken down by browser for all sites on data.gov.
data = requests.get("https://analytics.usa.gov/data/live/browsers.json")
print("90 days of visits broken down by browser for all sites:")
print(data.json()['totals']['browser'])
```

```python
#4.	Write a Python program to extract and display all the image links from en.wikipedia.org/wiki/Peter_Jeffrey_(RAAF_officer)
import re
url = "https://en.wikipedia.org/wiki/Peter_Jeffrey_(RAAF_officer)"
data = requests.get(url).text
img_obj = BeautifulSoup(data,"html5lib")
for link in img_obj.find_all('img', {'src':re.compile('.jpg')}):
    print(link['src']+'\n')
```

```python
#5.	Write a Python program to get the number of people visiting a U.S. government website right now. Source: https://analytics.usa.gov/data/live/realtime.json
url = ' https://analytics.usa.gov/data/live/realtime.json'
us_obj = requests.get(url).json()
print("Number of people visiting a U.S. government website right now  is :")
print(us_obj['data'][0]['active_visitors'])
```

```python
#8.	Write a Python program to display the contains of different attributes like different attributes like status_code, headers, url, history, encoding, reason, cookies, elapsed, request and content of a specified resource.
response = requests.get('https://npowercanada.ca/')

print("Status Code: ",response.status_code)
print("Headers: ",response.headers)
print("Url: ",response.url)
print("History: ",response.history)
print("Encoding: ",response.encoding)
print("Reason: ",response.reason)
print("Cookies: ",response.cookies)
print("Elapsed: ",response.elapsed)
print("Request: ",response.request)
print("Content: ",response._content)
```
