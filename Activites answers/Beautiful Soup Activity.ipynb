```python
# 1. Write a Python program to find the title tags from a given html document.
import requests
from bs4 import BeautifulSoup

# doc_url = 'https://drive.google.com/file/d/1jAq0sJqBRquog1RDN_bCQmUfM0Z56MOs/view?usp=sharing'
# resp = requests.get(doc_url)
# html_content = resp.text

with open("BeautifulSoup.html", "r") as f:
    html_content = f.read()

soup = BeautifulSoup(html_content, 'html.parser')

title_tags = soup.find_all('title')

# Print the title tags
print(title_tags)
```

```python
# 2. Write a Python program to get the number of paragraph tags of a given html document.
paragraph_tags_count = len(soup.find_all('p'))
print("Number of paragraph tags:", paragraph_tags_count)
```

```python
# 3. Write a Python program to extract the text in the first paragraph tag of a given html document.
first_paragraph_text = soup.find('p').text
print("Text in the first paragraph tag:", first_paragraph_text)
```

```python
# 4. Write a Python program to extract all the URLs from the webpage python.org that are nested within <li> tags from

import requests

base_url = 'https://python.org'
response = requests.get(base_url)
html_content = response.text

soup = BeautifulSoup(html_content, 'html.parser')

# urls = [li.a['href'] for li in soup.find_all('li') if li.a]
# print("URLs within <li> tags:", *urls, sep='\n')

urls = soup.find_all('a', href=True)
print('Total urls: ', len(urls))

list_items = soup.find_all('li')
print('Total <li> elements: ', len(list_items))

lis_a = set()

for list_item in list_items:
    nested_a = list_item.find_all('a', href=True)
    lis_a.update([a['href'] for a in nested_a if '/' in a.get('href', '')])

print('Total nested <a>:', len(lis_a))
print(*lis_a, sep='\n')
```

```python
# 5. Write a Python program to extract all the text from a given web page.

with open("BeautifulSoup.html", "r") as f:
    html_content = f.read()

soup = BeautifulSoup(html_content, 'html.parser')

all_text = soup.get_text()
print("All text from the webpage:", all_text)
```

```python
# 6. Write a Python program to retrieve the HTML code of the title, its text, and the HTML code of its parent.

with open("BeautifulSoup.html", "r") as f:
    html_content = f.read()

soup = BeautifulSoup(html_content, 'html.parser')

title_tag = soup.title
print("HTML code of the title:", title_tag)
print("Text of the title:", title_tag.text)
print("HTML code of the parent of the title:", title_tag.parent)
```

```python
# 8. Write a Python program to remove the contents of a tag in a given html document.

with open("BeautifulSoup.html", "r") as f:
    html_content = f.read()

soup = BeautifulSoup(html_content, 'html.parser')
tag_to_empty = 'h2'
soup.find(tag_to_empty).clear()

print(soup)
```

```python
# 9.	Write a Python program to wrap an element in the specified tag and create the new wrapper.

with open("BeautifulSoup.html", "r") as f:
    html_content = f.read()

soup = BeautifulSoup(html_content, 'html.parser')

element_to_wrap = 'h2'
new_wrapper_tag = 'div'

new_wrapper = soup.new_tag(new_wrapper_tag)
soup.find(element_to_wrap).wrap(new_wrapper)

print(soup)
```

```python
# 10. Write a Python program to remove a tag from a given tree of html document and destroy it and its contents.

with open("BeautifulSoup.html", "r") as f:
    html_content = f.read()

soup = BeautifulSoup(html_content, 'html.parser')

tag_to_remove = 'h2'

soup.find(tag_to_remove).decompose()

print(soup)
```
