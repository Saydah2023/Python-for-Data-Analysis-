```python
#4. What are the mean and standard deviation of age for those who earn more than
#50K per year (salary feature) and those who earn less than 50K per year?

mean_age_50k_more = df[df['income'] == '>50K']['age'].mean()
print("The mean value for persons who earn 50k and more is:" ,mean_age_50k_more)

std_age_50k_more = df[df['income'] == '>50K']['age'].std()
print("The standard deviation value for persons who earn 50k and more is:" ,std_age_50k_more)

mean_age_50k_less = df[df['income'] == '<=50K']['age'].mean()
print("The mean value for persons who earn less than 50K is:" ,mean_age_50k_less)

std_age_50k_less = df[df['income'] == '<=50K']['age'].std()
print("The standard deviation value for persons who earn less than 50K is:" ,std_age_50k_less)
```

```python
#5. Is it true that people who earn more than 50K have at least a high school
#education? (education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc,
#Masters or Doctorate feature)

high_education = ['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate']
high_education_50k_more = df[df['income'] == '>50K']['education'].isin(high_education).all()
print("\nDo people who earn >50K have at least a high school education?", high_education_50k_more)

df[df['income']=='>50K']['education'].value_counts()

#df[df['income']== '>50K']['education'].value_counts()
```

```python
#7. Among whom is the proportion of those who earn a lot (>50K) greater: married or
#single men (marital-status feature)? Consider as married those who have
#a marital-status starting with Married (Married-civ-spouse, Married-spouse-absent
#or Married-AF-spouse), the rest are considered bachelors

'''df.loc[(df['income'] >'50K') &
     (df['marital-status'].isin(['Never-married', 
                                   'Separated', 
                                   'Divorced',
                                   'Widowed'])), 'income'].value_counts()
'''
married = df['marital-status'].str.startswith('Married')
proportion_50k_more_married = (df[(df['gender'] == 'Male') & married]['income'] == '>50K').mean()
proportion_50k_more_single = (df[(df['gender'] == 'Male') & ~married]['income'] == '>50K').mean()
print("\nProportion of married men earning >50K:", proportion_50k_more_married)
print("Proportion of single men earning >50K:", proportion_50k_more_single)
if proportion_50k_more_married > proportion_50k_more_single:
    print("Proportion of those who earn a lot (>50K) is greater among married men.")
else:
    print("Proportion of those who earn a lot (>50K) is greater among single men.")
```
